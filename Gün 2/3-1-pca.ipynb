```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
features = iris.feature_names
df = pd.DataFrame(x, columns=features)
df["species"] = y
```

```python
df
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(121, projection='3d')
ax.scatter(df['sepal length (cm)'], df['sepal width (cm)'], df['petal length (cm)'], c=df['species'], cmap='viridis')
ax.set_title('3D Scatter - PCA Ã–ncesi Veri')
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_zlabel('Petal Length (cm)')
```

```python
# normalizasyon bunu sonra anlatacaÄŸÄ±z
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
```

```python
# dataframe e dÃ¶nÃ¼ÅŸtÃ¼relim
df_pca = pd.DataFrame(x_pca, columns=["PC1","PC2"])
df_pca["species"] = y
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df_pca['species'], cmap='viridis')
plt.title('PCA - Iris Veri Seti')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar(label='Species')
plt.show()
```
