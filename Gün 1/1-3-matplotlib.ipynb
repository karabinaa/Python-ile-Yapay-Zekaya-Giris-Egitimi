```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = x ** 2 
y
```

```python
plt.plot(x,y)
plt.title("1:10 aralığında x² fonksiyonun grafiği")
plt.ylabel("Sayının karesi")
plt.xlabel("Sayı")
plt.show()
```

```python
x = np.linspace(0,2*np.pi, 100)
y1, y2 = np.sin(x),np.cos(x)

plt.plot(x,y1, label = 'sinx', color = 'b')
plt.plot(x,y2, label = 'cosx', color = 'r', linestyle='--')

plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
plt.title('sinx ve cosx eğrileri tek plotta')

plt.legend()
plt.show()
```
